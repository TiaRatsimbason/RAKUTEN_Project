```python
import pandas as pd
import numpy as np
import os
import shutil
```

# X_TEST

```python
src_dir = "../data/preprocessed/"
dst_dir = "../data/container/"

csv_file = "X_test_update.csv"

df = pd.read_csv(f"{src_dir}{csv_file}")
df = df.head(1000)
df.to_csv(f"{dst_dir}{csv_file}")
df

src_dir = "../data/preprocessed/image_test"
dst_dir = "../data/container/image_test"


os.makedirs(dst_dir, exist_ok=True)

for _, row in df.iterrows():
    imageid = row["imageid"]
    productid = row["productid"]
    name_file = f"image_{imageid}_product_{productid}.jpg"
    src_path = os.path.join(src_dir, name_file)
    dst_path = os.path.join(dst_dir, name_file)

    if os.path.exists(src_path):
        shutil.copy(src_path, dst_path)
        print(f"Copied {name_file} to {dst_path}")
    else:
        print(f"Source file {name_file} does not exist.")
```
